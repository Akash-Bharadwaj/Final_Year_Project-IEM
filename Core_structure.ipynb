## Training The Machine

```python
df1 = pd.read_excel("C:\\Akash\\final_year_project\\Core_Dataframe.xlsx")
df1
```

### Data Processing

```python
df1['Driving style'] = df1['Driving style'].apply(lambda x : 1 if x=='Calm' else 0)
df1['Vehicle moving alongside'] = df1['Vehicle moving alongside'].apply(lambda x : 1 if x=='Yes' else 0)
df1['diversion way available'] = df1['diversion way available'].apply(lambda x : 1 if x=='Yes' else 0)
df1['vehicle side'] = df1['vehicle side'].apply(lambda x : 1 if x=='left-left' else 2 if x=='left-right' else 3 if x=='right-left' else 4)
df1['Want to turn'] = df1['Want to turn'].apply(lambda x : 1 if x=='right' else 2 if x=='left' else 0)
```

```python
df1
```

### Model Building and Training the data

```python
x = df1.drop('Decision Taken by a Driver',axis='columns')
y = df1["Decision Taken by a Driver"]
```

```python
y = y.apply(lambda x : 1 if x=='drive straight and maintain same speed' 
                    else 2 if x=='drive straight and increase speed' 
                    else 3 if x=='slow down' 
                    else 4 if x=='Turn right increase speed Turn Left (Overtake)' 
                    else 5 if x=='Turn left increase speed Turn right (Overtake)' 
                    else 6 if x=='Brake' 
                    else 7 if x=='Emergency Brake' 
                    else 8 if x=='Give pass to the vehicle behind ' 
                    else 9 if x=='Turn Left' 
                    else 10 if x=='Turn Right'
                    else 11 if x=='slow down take a side and stop'
                    else 0)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

## Hyper Parameter Tuning using GridSearchCV

```python
model_parameter = {
    'linear_regression': {
        'model' : LinearRegression(),
        'params': {
            'normalize' : [True , False]
        }
    },
    'lasso':{
        'model': Lasso(),
        'params': {
            'alpha' : [1,2],
            'selection' : ['random' , 'cyclic']
        }
    },
    'decision_tree':{
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion' : ['mse','friedman_mse'],
            'splitter' : ['best','random']
        }
    }

}
```

```python
scores = []
cv = ShuffleSplit(n_splits=5 , test_size=0.2 , random_state=0)
for model_name , mp in model_parameter.items():
    clf = GridSearchCV(mp['model'] ,mp['params'], cv=cv, return_train_score=False)
    clf.fit(x,y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_paramater': clf.best_params_
    })

dfr = pd.DataFrame(scores,columns=['model','best_score','best_paramater'])
dfr
```

```python
final_model = DecisionTreeRegressor()
final_model.fit(x_train , y_train)
```

## Functions for Prediction

```python
def predict_move(Driving_style,vehicle_side,Front_way_clear_upto_distance,Vehicle_moving_alongside,Vehicle_alongside_speed,Vehicle_in_front_speed,Vehicle_in_front_side_clear_upto_distance,Vehicle_approaching_from_behind_speed,Vehicle_approaching_from_behind_distance,Response_Time,Want_to_turn,diversion_way_available):
    c = np.zeros(len(x.columns))

    if Driving_style == 'Calm':
        c[0] = 1
    else :
        c[0] = 0
    
    if vehicle_side == 'left-left':
        c[1] = 1
    elif vehicle_side == 'left-right':
        c[1] = 2
    elif vehicle_side == 'right-left':
        c[1] = 3
    else :
        c[1] = 4

    c[2] = Front_way_clear_upto_distance

    
    if Vehicle_moving_alongside == 'Yes':
        c[3] = 1
    else:
        c[3] = 0

    c[4] = Vehicle_alongside_speed
    c[5] = Vehicle_in_front_speed
    c[6] = Vehicle_in_front_side_clear_upto_distance
    c[7] = Vehicle_approaching_from_behind_speed
    c[8] = Vehicle_approaching_from_behind_distance
    c[9] = Response_Time
    
    if Want_to_turn == 'right':
        c[10] = 1
    elif Want_to_turn == 'left':
        c[10] = 2
    else :
        c[10] = 0
    
    if diversion_way_available == 'Yes':
        c[11] = 1
    else :
        c[11] = 0


    ans = final_model.predict([c])[0]

    if ans == 1:
        return 'drive straight and maintain same speed'
    elif ans == 2:
        return 'drive straight and increase speed'
    elif ans == 3:
        return 'slow down'
    elif ans == 4:
        return 'Turn right increase speed Turn Left (Overtake)'
    elif ans == 5:
        return 'Turn left increase speed Turn Left (Overtake)'
    elif ans == 6:
        return 'Brake'
    elif ans == 7:
        return 'Emergency Brake'
    elif ans == 8:
        return 'Give pass to the vehicle behind '
    elif ans == 9:
        return 'Turn Left'
    elif ans == 10:
        return 'Turn Right'
    elif ans == 11:
        return 'slow down take a side and stop'
    else :
        return 0

```
